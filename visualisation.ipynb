```python
x=data.sepallength
plt.hist(x,bins=10,color='orange',alpha=0.7,edgecolor='black')
plt.axvline(x.mean(),color='r',linestyle='dotted',linewidth=2)#mean of sepallength
plt.axvline(x.median(),color='g',linestyle='dashed',linewidth=2)#median of sepallength
plt.xlabel('values')
plt.ylabel('frequancy')
plt.title('Histogram of sepallength')
plt.show()
```

```python
plt.plot(x)
plt.xlabel('values')
plt.ylabel('frequancy')
plt.title('Line Graph for sepallength')
plt.show()
```
